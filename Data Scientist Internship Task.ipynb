### 1) Classify the options data into various expiries dates. Various expiry dates are mentioned as 04Mar21, 01Apr21, 11Mar21 and so on. Every Thursday there is an expiry.

```python
df['Date'] = df['Date'].astype(str) +" "+ df["Time"].astype(str)
```

```python
# Converting Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Creating a list of expiry dates
expiry_dates = []
for i in range(4):
    thursday = df['Date'].min() + pd.DateOffset(days=(3-df['Date'].min().weekday()+7*i))
    expiry_dates.append(thursday.strftime('%d%b%y').upper())
```

```python
# Creating a dictionary to store options data by expiry date
expiry_data = {}
for date in expiry_dates:
    expiry_data[date] = {}
```

```python
# Looping through each row in the dataframe and add options data to the corresponding expiry date
for index, row in df.iterrows():
    date = row['Date']
    expiry = (date + pd.DateOffset(days=(3-date.weekday()))).strftime('%d%b%y').upper()
    ticker = row['Ticker']
    open_interest = row['Open Interest']
    if expiry in expiry_dates:
        if ticker not in expiry_data[expiry]:
            expiry_data[expiry][ticker] = {'Open Interest': 0}
        expiry_data[expiry][ticker]['Open Interest'] += open_interest
```

```python
# Printing the expiry data
for expiry, data in expiry_data.items():
    print(expiry)
    print("-" * 20)
    for ticker, ticker_data in data.items():
        open_interest = ticker_data['Open Interest']
        print(f"{ticker}: {open_interest}")
    print("=" * 20)
```

### 2) The current data is for 1 minute interval. Make it in a way that it is available at 3 minutes, 5 minutes, 15 minutes & 60 minutes interval.

```python
df.head()
```

```python
# df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
# Define a list of resampling intervals
resample_intervals = ['3T', '5T', '15T', '60T']
```

```python
# Resample the data for each interval and calculate OHLCV and open interest data
resampled_data = {}
for interval in resample_intervals:
    resampled_df = df.resample(interval).agg({
        'Open': 'first',
        'High': 'max',
        'Low': 'min',
        'Close': 'last',
        'Volume': 'sum',
        'Open Interest': 'last'
    })
    resampled_data[interval] = resampled_df.dropna()
```

```python
# Print the resampled data for each interval
for interval, data in resampled_data.items():
    print(interval)
    print("-" * 77)
    print(data)
    print("=" * 30)
```

```python
import py_vollib.black_scholes.implied_volatility as iv
```

```python
# Filtering the data to only include options data
# options_data = df[df['Ticker'].str.contains('BANKNIFTY')]
```

```python
# Calculate the Future price based on the last Close price of BANKNIFTY
future_price = df[df['Ticker'] == 'BANKNIFTY01APR2130600CE.NFO']['Close'].iloc[-1]
```

```python
# Function to calculate the implied volatility
def get_implied_volatility(row):
    try:
        # Calculate the option price using Black Scholes model
        option_price = iv.implied_volatility_of_undiscounted_option_price(row['Close'], future_price, row['Strike'], row['Days_to_expiry']/365, row['Interest_Rate']/100, row['Option_Type'].lower())

        return option_price
    except:
        return np.nan
```

```python
# Apply the function to each row in the dataset
df['Implied_Volatility'] = df.apply(get_implied_volatility, axis=1)

# Print the dataset with the implied volatility
print(df)
```
